```python
#문제 1번

#제시된 단어들 중 랜덤으로 나온 단어로 3행시를 짓는 프로그램을 만드시오. (단, 아래의 조건이 모두 충족되어야 함.)


#['유재석', '박명수', '정준하', '정형돈', '노홍철', '하동훈']


#보기에 제시된 단어가 랜덤 리스트에 다 들어가야 함. 더 추가되거나 없으면 안됨.
#랜덤으로 뽑인 단어의 음절마다 각각 입력값을 받아야 함.
#제시된 각각의 음절과 입력값을 첫 글자가 같지 않다면, 아래와 같은 문구가 각각 나와야 함.
#"뽑힌 단어의 n번째 글자와 입력한 시작 글자가 다릅니다. 알맞게 입력해 주세요."
#<출력값 예시>

#랜덤으로 뽑힌 단어는 '박명수' 입니다.
#박 ! 박식하기 그지없고
#명 ! 명석한 두뇌를 가진
#수 ! 명수옹
#뽑힌 단어의 세번째 글자와 입력한 시작 글자가 다릅니다. 알맞게 입력해 주세요.
```

```python
#장수정님 정답

word_list = ['유재석', '박명수', '정준하', '정형돈', '노홍철', '하동훈']

import random
word = random.choice(word_list)

print("랜덤으로 뽑힌 단어는 '{}' 입니다.".format(word))

word1 = input('{} ! '.format(word[0]))
word2 = input('{} ! '.format(word[1]))
word3 = input('{} ! '.format(word[2]))

if str(word[0]) != str(word1[0]): 
    print("뽑힌 단어의 첫번째 글자와 입력한 시작 글자가 다릅니다. 알맞게 입력해 주세요.")
if str(word[1]) != str(word2[0]):
    print("뽑힌 단어의 두번째 글자와 입력한 시작 글자가 다릅니다. 알맞게 입력해 주세요.")
if str(word[2]) != str(word3[0]):
    print("뽑힌 단어의 세번째 글자와 입력한 시작 글자가 다릅니다. 알맞게 입력해 주세요.")
```

```python
#정답을 본뒤에 내가 원하느대로 살짝 바꿔봄
#위에는 잘못 입력한결과를 전부 나중에 볼수있는데 실시간으로 틀리게 입력했는지 볼수있고
#다시 입력할수 있게 만들었음

word_list = ['유재석', '박명수', '정준하', '정형돈', '노홍철', '하동훈']

import random
word = random.choice(word_list)

word1 = " "
word2 = " "
word3 = " "
print("랜덤으로 뽑힌 사람은 '%s' 입니다" % word)


while word1[0] != word[0]:
    print(word[0])
    word1 = input()
    
    if word1[0] == word[0]:
        break
    
    else:
        print("다시입력해주세요")

while word2[0] != word[1]:
    print(word[1])
    word2 = input("")
    
    if word2[0] == word[1]:
        break
    
    else:
        print("다시입력해주세요")

while word3[0] != word[2]:
    print(word[2])
    word3 = input("")
    
    if word3[0] == word[2]:
        break
    
    else:
        print("다시입력해주세요")
        
print("완성된 3행시 입니다")
print(word[0]," : ",word1)
print(word[1]," : ",word2)
print(word[2]," : ",word3)

        

    
    



    


    








```
